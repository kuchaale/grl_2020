```python
import glob
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import shapely.geometry as sgeom
from matplotlib.ticker import FuncFormatter

%matplotlib inline
```

```python
timescale = 20
DJF_bool = '_DJFonly'
cesta = '/mnt/4data/CMAM/0A.daily/composites_woSSW{}/'.format(DJF_bool)
```

# Data loading

## Tropopause

```python
infile_name = '{}{}_{}_comp_*_{}days.nc'.format(cesta, 'ptrop', 'absolute', 20)
infiles = glob.glob(infile_name)
e_types = list(map(lambda x: x.split('comp_')[1].split('_{}days'.format(timescale))[0], infiles))
da_trop =  xr.open_mfdataset(infiles, concat_dim='e_type', combine = 'nested')
da_trop['e_type'] = e_types
da_trop['ptrop'] = da_trop.ptrop/100.
da_trop
```

## LWA

```python
var_ls = ['lwatend_logH']
e_type_ls = ['himalayas', 'eastasia', 'westamer']
what_ls = ['anomalies','absolute']
da_ls2 = []

for what in what_ls:
    da_ls = []
    print(what)
    for i,e_type in enumerate(e_type_ls):
        print(e_type)
        ds_ls = []
        for var in var_ls:
            ds = xr.open_dataset(f'{cesta}{var}_{what}_comp_{e_type}_{timescale}days.nc')
            if var not in ['lwa','sink', 'lwatend_logH']:
                ds['plev'] = ds.plev/100.
            else:
                ds = ds.rename({'z': 'plev'})
            ds_ls.append(ds)
        da = xr.merge(ds_ls)

        da_ls.append(da)
    da_conc = xr.concat(da_ls, dim = 'e_type')
    da_conc['e_type'] = e_type_ls
    da_ls2.append(da_conc)
        
da_all = xr.concat(da_ls2, dim = 'what')
da_all['what'] = what_ls
da_all
```

## Stat. significance

```python
#infile_name = '{3}{0}_pvalues_from{1}_comp_*_{2}days_zm.nc'.format(var, 10000, timescale, cesta)
var = 'lwatend'#'sink'
infile_name = '{3}{0}_pvalues_from{1}_comp_*_{2}days.nc'.format(var, 10000, timescale, cesta)
print(infile_name)
infiles = glob.glob(infile_name)
comp_types = list(map(lambda x: x.split('comp_')[1].split('_{}days'.format(timescale))[0], infiles))
print(comp_types)
print(infiles)
print()
ds_p = xr.open_mfdataset(infiles, concat_dim='e_type', combine ='nested')
ds_p['e_type'] = comp_types
if var not in ['lwa','sink','lwatend']:
    ds_p['plev'] = ds_p.plev/100.
else:
    ds_p = ds_p.rename({'z': 'plev'})
ds_p
```

# Figure 4

```python
levels= np.linspace(-2.5,2.5,41)
cbar_label = 'sink [m/s/day]'
titles = ['Himalayas', 'East Asia', 'Rocky Mountains'] #'West America']
#var = 'vmro3'
#anomalies
mpl.rcParams.update({'font.size': 18})
scale = 24*3600.
lat = da_all.lat.values
cmap = mpl.cm.PuOr#RdBu_r
p = (da_all['lwatend'].sel(what = what_ls[0], lag = 0, plev = slice(None,60000)).mean('lon')*scale).plot.contourf(x= 'lat', \
                                                               y = 'plev', \
                                                               col = 'e_type', \
                                                               size = 5, \
                                                               cmap = cmap, levels = levels,\
                                                               add_colorbar = False)


p.data = da_all['lwatend'].sel(what = what_ls[1], lag = 0, plev = slice(None,60000)).mean('lon')*scale
levels2 = [0.5,1]
lev0 = [0]
    
plot_kwargs = dict(\
                   colors = 'grey', \
                   levels=levels2, add_colorbar=False, alpha =1 ) #0.7 #
p.map_dataarray(xr.plot.contour, 'lat', 'plev', **plot_kwargs)


plot_kwargs['linestyles'] = 'dashed'
plot_kwargs['levels'] = np.array(levels2)[::-1]*-1
p.map_dataarray(xr.plot.contour, 'lat', 'plev', **plot_kwargs)
plot_kwargs['linestyles'] = 'solid'
plot_kwargs['levels'] = lev0
plot_kwargs['linewidths'] = 3
plot_kwargs['alpha'] = 1

p.map_dataarray(xr.plot.contour, 'lat', 'plev', **plot_kwargs)




sel_dict = dict(plev = slice(None, 60000), lat = slice(-20,None), lag = 0)
p.data  = ds_p['lwatend'].sel(**sel_dict)
plot_kwargs2 = dict(levels = [0,0.05], hatches=['\\\\',None], \
                                colors='none', add_colorbar=False, \
                                 yincrease = False)
p.map_dataarray(xr.plot.contourf, 'lat', 'plev', **plot_kwargs2)

plot_kwargs2['levels'] = [0,0.01]
plot_kwargs2['hatches'] = ['////',None]
p.map_dataarray(xr.plot.contourf, 'lat', 'plev', **plot_kwargs2)


p.axes[0][0].set_ylabel('height [km]', fontsize = 19)
p.axes[0][0].set_xlabel('')
p.axes[0][1].set_xlabel('latitude [degree]', fontsize = 19)
p.axes[0][2].set_xlabel('')

p0=1013.25
H = 7000
da_trop2 = da_trop.ptrop.sel(lag=0, lat = slice(0,None)).mean('lon')
label_ls = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
for i,(ax, sel_dict, label) in enumerate(zip(p.axes.flat, p.name_dicts.flat, label_ls)):
    ax.text(0.05, 0.95, label, transform=ax.transAxes, fontsize = 20, 
              fontweight='bold', va='top', bbox={'alpha': 0.7, 'facecolor': 'white'})
    print(i)
    
    e_type = sel_dict['e_type']
    if e_type == 'himalayas':
        box = sgeom.box(minx=70, maxx=105, miny=20, maxy=42)
    elif e_type == 'eastasia':
        box = sgeom.box(minx=110, maxx=145, miny=27.5, maxy=52)
    elif e_type == 'westamer':
        box = sgeom.box(minx=-125, maxx=-102.5, miny=27.5, maxy=52)
    
    
    ax.axis([0, 90, 0, 60000])
    
    ax.set_yticks([10000,30000,50000])
    #ax.set_title('')#
    ax.set_title(titles[i], fontdict=dict(fontweight = 'bold', fontsize = 22))
    fmt = mpl.ticker.FormatStrFormatter("%g")
    def thousands(x, pos):
        return '%g' % (x*1e-3)

    fmt = FuncFormatter(thousands)
    ax.yaxis.set_major_formatter(fmt)
    ax.set_xticks([0,0,20,40,60,80])
    
    ax2 = ax.twinx()
    ax2.get_shared_y_axes()
    ax2.set_yscale('log')
    ax2.axis([0, 90, p0*np.exp(-0/H), p0*np.exp(-60000/H)])
    ax2.set_yticks([1000,100,10,1])   
    ax2.yaxis.set_major_formatter(fmt)
    
    ax2.tick_params(direction='out', which='both', labelsize=14)
    ax2.hlines(70,-20,90, color = 'green', linestyles='--', linewidths = 3)
    ax2.vlines(list(box.bounds)[1::2],90,50, color = 'green', linestyles='--', linewidths = 3)
    #ax2.plot(lat, da_trop2.sel(**sel_dict), 'k--', lw = 3)
    da_trop2.sel(**sel_dict).plot.line('k--', ax = ax2, lw = 3,)
    ax2.set_title('')
    ax2.set_ylabel('')
    
    if i != 2:
        ax2.set_yticklabels([])
    if i == 2:
        ax2.set_ylabel('pressure [hPa]', fontsize = 19)


p.fig.subplots_adjust(right=0.875)
cbar_ax = p.fig.add_axes([0.075, -0.03, .8, .03])
norm = mpl.colors.BoundaryNorm(levels[1:-1], cmap.N)
label = cbar_label
cb2 = mpl.colorbar.ColorbarBase(cbar_ax, cmap=cmap,  \
                                norm=norm, orientation='horizontal', \
                                boundaries=levels, extend='both',  label = cbar_label, \
                                ticks=[-2,-1,0,1,2], spacing='proportional')

outfile='{}_{}_{}_{}days_zm_wsignificance_DJFonly.pdf'.format('lwatend', what_ls[0], 'all', timescale)
print(outfile)
plt.savefig(outfile, dpi = 300, bbox_inches='tight')
```
